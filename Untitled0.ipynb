```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.stats.api as sms
from scipy.stats import shapiro

# =====================================================
# 1. MEMBACA DATASET
# =====================================================
file_path = 'coffee_shop_revenue.csv'
df = pd.read_csv(file_path)

print("="*60)
print("INFORMASI DATASET")
print("="*60)
print(df.info())

print("\n5 DATA PERTAMA")
print(df.head())

# =====================================================
# 2. MENENTUKAN VARIABEL
# =====================================================

# Variabel Bebas (X)
X = df[
    [
        'Number_of_Customers_Per_Day',
        'Average_Order_Value',
        'Operating_Hours_Per_Day',
        'Number_of_Employees',
        'Marketing_Spend_Per_Day',
        'Location_Foot_Traffic'
    ]
]

# Variabel Terikat (Y)
Y = df['Daily_Revenue']

# Menambahkan konstanta
X_with_const = sm.add_constant(X)

# =====================================================
# 3. MEMBUAT MODEL REGRESI LINEAR BERGANDA
# =====================================================
model = sm.OLS(Y, X_with_const).fit()

print("\n" + "="*60)
print("HASIL REGRESI LINEAR BERGANDA")
print("="*60)

print(model.summary())

# =====================================================
# 4. UJI ASUMSI KLASIK
# =====================================================

print("\n" + "="*60)
print("HASIL UJI ASUMSI KLASIK")
print("="*60)

# -----------------------------------------------------
# A. UJI MULTIKOLINEARITAS (VIF)
# -----------------------------------------------------
vif_data = pd.DataFrame()
vif_data["Variabel"] = X_with_const.columns

vif_data["VIF"] = [
    variance_inflation_factor(
        X_with_const.values, i
    )
    for i in range(len(X_with_const.columns))
]

print("\n[1] UJI MULTIKOLINEARITAS")
print(vif_data.iloc[1:])

# -----------------------------------------------------
# B. UJI NORMALITAS RESIDU
# -----------------------------------------------------
residuals = model.resid

shapiro_test = shapiro(residuals)

print("\n[2] UJI NORMALITAS RESIDU (SHAPIRO-WILK)")
print(f"Statistik = {shapiro_test.statistic:.4f}")
print(f"P-Value   = {shapiro_test.pvalue:.4f}")

if shapiro_test.pvalue > 0.05:
    print("Kesimpulan: Residu Berdistribusi Normal")
else:
    print("Kesimpulan: Residu Tidak Berdistribusi Normal")

# -----------------------------------------------------
# C. UJI HETEROSKEDASTISITAS
# -----------------------------------------------------
bp_test = sms.het_breuschpagan(
    residuals,
    X_with_const
)

print("\n[3] UJI HETEROSKEDASTISITAS (BREUSCH-PAGAN)")
print(f"P-Value = {bp_test[1]:.4f}")

if bp_test[1] > 0.05:
    print("Kesimpulan: Tidak Terjadi Heteroskedastisitas")
else:
    print("Kesimpulan: Terjadi Heteroskedastisitas")

# -----------------------------------------------------
# D. UJI AUTOKORELASI
# -----------------------------------------------------
dw = sm.stats.durbin_watson(residuals)

print("\n[4] UJI AUTOKORELASI (DURBIN-WATSON)")
print(f"Nilai DW = {dw:.4f}")

if 1.5 <= dw <= 2.5:
    print("Kesimpulan: Tidak Ada Autokorelasi")
else:
    print("Kesimpulan: Ada Indikasi Autokorelasi")

# =====================================================
# 5. PREDIKSI
# =====================================================

df["Prediksi_Revenue"] = model.predict(X_with_const)

print("\n5 HASIL PREDIKSI PERTAMA")
print(df[["Daily_Revenue", "Prediksi_Revenue"]].head())

# =====================================================
# 6. VISUALISASI
# =====================================================

plt.figure(figsize=(14,5))

# -----------------------------------------------------
# Grafik 1 : Aktual vs Prediksi
# -----------------------------------------------------
plt.subplot(1,2,1)

plt.scatter(
    Y,
    model.fittedvalues,
    alpha=0.7,
    edgecolors='black'
)

plt.plot(
    [Y.min(), Y.max()],
    [Y.min(), Y.max()],
    'r--'
)

plt.title("Aktual vs Prediksi Penjualan")
plt.xlabel("Penjualan Aktual")
plt.ylabel("Penjualan Prediksi")
plt.grid(True)

# -----------------------------------------------------
# Grafik 2 : Residual Plot
# -----------------------------------------------------
plt.subplot(1,2,2)

plt.scatter(
    model.fittedvalues,
    residuals,
    alpha=0.7,
    edgecolors='black'
)

plt.axhline(
    y=0,
    color='red',
    linestyle='--'
)

plt.title("Residual Plot")
plt.xlabel("Nilai Prediksi")
plt.ylabel("Residual")
plt.grid(True)

plt.tight_layout()

plt.savefig(
    "hasil_regresi_penjualan.png",
    dpi=300
)

plt.show()
```
